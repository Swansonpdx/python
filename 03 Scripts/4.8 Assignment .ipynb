```python
df_merged_both['spending'] = df_merged_both.groupby(['user_id'])['prices'].transform(np.mean)

df_merged_both.loc[df_merged_both['spending'] < 10, 'spending_flag'] = 'Low Spender'
df_merged_both.loc[df_merged_both['spending'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
df_merged_both[['user_id', 'spending_flag', 'spending']].head(10)
```

```python
#7

df_merged_both['frequency'] = df_merged_both.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_merged_both.loc[df_merged_both['frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_merged_both.loc[(df_merged_both['frequency'] <= 20) & (df_merged_both['frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
df_merged_both.loc[df_merged_both['frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
